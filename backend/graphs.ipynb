```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
top5_asileXuser = pd.read_csv("top5_aisleXuser.csv")
```

```python
aisle_counts = top5_asileXuser['aisle'].value_counts().sort_values(ascending=False)

# Opcional: mostrar solo los 20 más frecuentes
aisle_counts = aisle_counts.head(20)

# Crear gráfico de barras
plt.figure(figsize=(12, 6))
aisle_counts.plot(kind='bar')
plt.title('Frecuencia de aparición por Aisle')
plt.xlabel('Aisle')
plt.ylabel('Cantidad de apariciones')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
